# 인공지능 - 독후감

## '빅데이터를 지배하는 통계의 힘'을 읽고

### 정보통신공학
### 1731061007 이대희

나는 인문계에서 이과를 나와서 대학교에서 공학사로 졸업하고, 현재는 정보통신전공으로 대학원을 다니고 있다. 거의 인생의 반을 IT/과학 분야에서 배우고, 듣고 살아왔다. 이런 나에게 통계라는 단어는 평소 많이 들어보며 생활했지만 조금은 멀게만 느껴지는 그런 생소한 분야였다. 어쩌면 내가 일상생활에서 심심찮게 접했던 대부분이 통계를 통해 나에게 정보로 다가왔었고, 나는 결코 그것을 중요하게 인식하지 않았던 것 같다.

최근 이슈로 대두되고 있는 인공지능, 빅데이터 등은 막대한 양의 데이터를 다루면서, 인간이 결코 지금까지 접하지 못했던 신세계를 보여주려 하고 있다. 이번에 읽은 ‘빅데이터를 지배하는 통계의 힘-통계학이 최강의 학문이다.’라는 책은 통계학의 중요성과 빅데이터에서 통계학의 필요성을 내세우고 있다. 제목만 봐도 통계학의 중요성을 강하게 어필하고 있다는 것이 느껴졌다. 

먼저 통계학은 어떤 분야에서는 데이터를 모아 분석해 가장 올바르고 빠른 답을 제시해준다. 대기업, 중소기업이든 하나의 기업에서 큰 프로젝트를 진행할 때, 통계를 이용하여 결정을 내려서 잘못된 판단일 때의 피해를 최소로 줄이려 하고 있다. 역사적으로는 영국에서 콜레라가 발병했을 때, 관련 데이터를 모아서 콜레라 발병 원인을 파악하고 피해를 막았었다. 교육학에서는 전국적으로 교육 관련 정책을 실행하기 전에 시범으로 소수의 학교에 정책을 도입하여 그 결과를 도출한다. 그리고 결과에 따른 통계학적 결론을 이용하여 정책이 적합한지를 따져본다. 더 나아가 심리학이나 사회학, 자연과학 분야에서도 가설을 검증할 때 적절한 데이터를 취하고 해석하기 위해 통계학의 지식을 이용한다. 이렇게 통계학은 분야를 막론하고 다양한 곳에서 최적의 결정을 내기 위해 사용되고 있다. 이런 이유로 이 책에서 통계학을 최강의 학문이라고 제시하고 있다. 처음에는 공대인으로서 이 말이 언짢았지만, 책의 중반부까지 읽고 나서 말의 뜻을 이해할 수 있었다.

책을 읽으면서 인상 깊었던 부분이 있었다. 통계학에서 데이터를 다루는 법 중에서 임의화 비교실험이 있다. 임의화 비교실험은 하나의 가설을 입증하기 위해 반복적으로 법칙을 발견하고, 이를 결론으로 도출해나가는 실험의 과정이다. 프랑스의 수학자이자 물리학자인 앙리 푸앵카레는 과학적 방법론의 중요한 특징을 ‘관찰과 실험’이라고 정의하였다. 앙리 푸앵카레는 이런 말을 하였다. ‘관찰이란 대상을 자세히 보고 측정해 그로부터 어떤 진실을 밝히는 행위이다. 실험은 다양하게 조건을 바꾼 상태에서 대상을 보고 측정해 그로부터 어떤 진실을 밝히는 행위이다.’ 그렇다. 비록 내가 석사과정 6개월 차 풋내기일지도 모르지만, 논문을 작성하는 과정이 이와 유사하다는 것을 뼈저리게 느꼈다. 내가 논문을 작성하려고 할 때 논문 주제를 정해야 한다. 기존의 방법보다 더 효율적이고 창의적인 주제를 정하는 이 과정이 관찰인 것이다. 그리고 주제를 정하게 되면 실험을 통해, 실험 환경을 변화시키면서 주제의 결론을 추출해가는 과정이 앙리 푸앵카레가 말한 실험이다. 그런데 바로 이 관찰과 실험 분야에서 통계학의 임의화 비교실험이 큰 힘을 발휘하게 된다. 내가 논문 주제를 정하고 실험을 함에 있어, 나에게는 주제를 완벽하게 입증할만한 데이터가 필요하다. 이 데이터를 얻기 위한 반복적인 실험을 통해 나의 가설을 입증해야만 하는 것이다.

빅데이터는 말 그대로 거대한 양의 데이터 트래픽을 관리하여 의미 있는 정보를 만들어 내는 과정이다. 이러한 막대한 양의 데이터를 이용하여 단 하나의 결정(결론)을 짓기 위해서는 데이터를 다루는 통계가 필요하다. 통계를 이용하면 다수의 결론 중에서 최선의 결론을 선택하여 원하고자 했던 결론과의 오차를 최소화시킬 수 있다. 빅데이터는 다양한 분야에서 사용될 것이다. 아니, 사용된다. 그럼으로써 이 어마어마한 양의 데이터 중에서 필요한 데이터를 취합하고 합산하는 과정이 얼마나 중요한 것인지를 이 책을 읽고 처음 느껴보았다. 이 책의 앞부분에서 저자는 ‘정보를 지배하는 자가 세계를 다스린다는 말을 현대에 맞게 바꾼다면, 통계학을 아는 자가 세계를 평정한다는 말이 된다’고 하였다.
