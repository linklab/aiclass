# 제목 : 빅데이터를 지배하는 통계의 힘
## 저자 : 니시우치 히로무
## 옮김 : 신현호

### 이상권 1731061008 정보통신공학과

빅데이터라는 용어는 해외 저널과 블로거들 사이에 사용됐다고 합니다. 빅데이터란 규모와 주기, 형식 등이 다양하고 복잡해 수집과 저장, 분석이 힘든 데이터를 말합니다. 사실 겉보기엔 쓸모없는 데이터로 보일 수 있지만 새로운 가치를 창조해 낸다는 의미에서 중요하게 여겨질 것 같다는 생각이 들었습니다. 미래 상황에 대한 결론을 예측할 수 있다고 저자는 말하며, 저 또한 이 주장이 틀리다고 생각하지 않습니다. 많은 국가 및 기업들이 빅데이터를 활용하기 위해 고분투구하고 있다고 합니다. 과연 이 쓸모없다고 느낄 수 있는 데이터들의 모음이 미래에 어떠한 결과를 불러 드릴지 매우 기대 합니다. 저자는 빅데이터가 기존의 난해한 의사결정을 도울 수 있다고 합니다. 가령 다수를 위해 소수를 희생시켜야 하는 상황에서 개인인 우리는 쉽게 결정을 할 수 있을까요? 이러한 질문의 이슈 하나하나들을 경험에서 비롯된 이야기로 서술하고 있습니다. 그리고 이를 통계학 관점에서 바라봅 니다. 통계학과 빅데이터는 얼핏 보아도 많은 연관성이 있습니다. 모든 과학적 분석 방법의 기본이라고 저자는 소개합니다. 어쩌면 데이터의 축적은 지금의 모든 이로운 가치들을 만들어냈다고도 생각이 듭니다. 거기다 빅데이터라는 방대한 경험들은 통계학과 부합되면서 다수의 가치와 동시에 밀도 높은 가치를 만들어 낼 수 있다고 생각합니다. 많은 주제 중 제가 가장 인상 깊게 본 주제는 ”오차와 인과관계가 통계학의 핵심이다“라는 장입니다. 특히 집계된 자료의 효율성을 따져보는 내용은 과거의 방식과 빅데이터를 통해 진화될 수 있는 현재의 방식을 세기별로 서술한 부분입니다. 간호사의 대표인물인 나이팅게일이 사용했던 통계 및 집계는 사망 확률을 높여주는 굉장한 사례라고 말합니다. 이렇듯 과거의 통계 및 집계는 지금의 의식(또는 인식), 생활, 건강, 편안함의 발전을 이루는데 가장 큰 공헌을 하지 않았나 생각합니다. 이러한 아날로그틱한 집계방식과 활용방법들은 지금의 시점에서 잘 부합되지 않는다고 저자는 말합니다. 이 주장에 어느 정도 동의하지만 완전히 필요 없다고는 생각하지 않습니다. 실제로 객관적인 자료나 통계는 평균적이고 보편화 한 결정에 도달합니다. 그러나 인간관계, 개인의 성향, 각 나라 및 지역의 의식, 환경 등을 고려해야할 필요는 있다고 생각합니다. 이러한 요소들 을 수식의 파라미터 형식으로 표현되고 실행되어 진다면 사람의 발전이 이루어질까요? ”보편적 의식을 기계가 정의한다.“는 것은 모든 인간이 똑같은 생각을 만들게 한다고 생각됩니다. 결코 좋지 못하다는 것을 알 수 있습니다. 마케팅에서의 빅테이터 활용은 기존의 시스템에 빅데이터라는 커다란 요소가 접목되면서 매우 중요한 가치로 인정받고 있습니다. 저자 또한 이익 향상에 인간관계를 생각하지 않는 통계해석은 의미가 없다고 말하며, 비구매자를 제와한 실구매자만을 대상으로 설문조사의 근거로 삼는다고 말 합니다. 빅데이터라는 단어가 나오기 전부터 언급됐던 내용임으로 사실 큰 의미가 없다고 생각합니다. 그러나 빅데이터와 통계학을 이야기 할 때 는 한 요소로도 볼 수 있다는 것에는 동의 합니다. 이러한 문제점들을 해결하기 위해 저자는 두 가지 솔루션을 제시합니다. 개인의 성격, 가정환경 등 관련된 조건을 계속적으로 추적하고 조사하여 통계학적 기법에 이용하자는 것과 통계의 조건을 최대한 공정하게 설정하는 것입니다. 이 두 가지는 완전히 다른 방향을 가졌다고 생각합니다. 후자는 앞서 서술한 본인이 느끼는 문제점과 동일하며, 절대 솔루션이 될 수 없다고 느껴집니다. 특정 상황에서 필요한 방법일 수는 있지만 절대 주가 돼서는 안 될 방법입니다. 그리고 첫 번째 솔루션은 이상적인 방법이라고 생각합니다. 다양한 조건을 고려해야하며, 많은 파라미터가 있어야 이러한 인과관계에서의 통계학이 의미 있다고 생각합니다.
